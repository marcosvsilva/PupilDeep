```python
import cv2
import csv
import time
import numpy as np
from matplotlib import pyplot as plt
```

```python
label_dataset = []
with open('C:\ProjetosMarcos\PupilDeep\dataset\labels.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=';')
    for row in csv_reader:
        label_dataset.append(row)
```

```python
def marc_point(img, center):
    yellow = (255,255,255)  
    cv2.circle(img, center, 5, yellow, 2)
    return img    

def rotate_img(img):
    img = cv2.rotate(img, cv2.ROTATE_180)
    return cv2.flip(img, 1)
```
